```python
movies_df = df[df['type'] == 'Movie'].copy()
movies_df.loc[:, 'duration'] = movies_df['duration'].str.replace(' min', '')
movies_df.loc[:, 'duration'] = pd.to_numeric(movies_df['duration'], errors='coerce')
movies_df = movies_df.dropna(subset=['duration'])
movies_df.loc[:, 'duration'] = movies_df['duration'].astype(float)
movies_df = movies_df.sort_values(by='duration', ascending=False)
top_10_longest_movies = movies_df.head(10)
result = top_10_longest_movies[['title', 'duration', 'director', 'country']]
result
```

```python
plt.figure(figsize=(6, 4))
plt.barh(result['title'], result['duration'], color='skyblue')
plt.xlabel('Duration (minutes)')
plt.ylabel('Movie Title')
plt.title('Top 10 Longest Duration Movies')
plt.gca().invert_yaxis() 
plt.show()
```
